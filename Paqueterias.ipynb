Vimos que podemos jalar una paqueteria existante y guerdarla a dentro de un objeto. Es posible tambien importar solo una parte de una paquetria.
En la ultima linea vieron que la paqueteria numpy esta escrita en sub-paqueteria como lo es **random**.
Se puede jalar unicamente la sub-paqueteria **random** de **numpy** de la manera siguente:

```python
from numpy import random as rd

rd.randn()
```

Es tambien util saber que muchas veces encontraran codigos donde se carga la paquetria sin ponerla dentro de un objeto. No me gusta pero es posible de cargar igualmente la sub-paqueteria **numpy.random** y usarla de la manera siguiente:

```python
from numpy import random

random.randn()
```

Veremos mas adelante que crear sus propias paqueterias es muy util para realizar grandes proyectos o para compartir sus funciones con los demas. Sin embargo, es importante aprender primero que son las funciones y como definirlas y usarlas.
